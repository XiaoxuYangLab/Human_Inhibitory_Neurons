```r
library(ggplot2)
library(dplyr)
library(gplots)
library(reshape2)
```

```r
filtered <- read.table("../7614_final_summary_anno3.txt",header=T,sep="\t")
variant_list <- read.csv("../variant_annotation/287_variant_annotation_bulk_sort_COUPTF2_TBR1_DLX1_CBH.csv",header=T,sep=",", row.names = 1)
```

```r
mosaic_filtered <- filtered[filtered$Mosaic_shared_btw_WGS_Amp==1,]
shared_variant_list<-variant_list[variant_list$sample_number_presented>1,]
shared_mosaic_filtered<-mosaic_filtered[mosaic_filtered$CHROM.POS.REF.ALT %in% rownames(shared_variant_list),]
```

```r
melt_mtcars<-shared_mosaic_filtered[,c("ID", "MAF", "CHROM.POS.REF.ALT","Cell_Type","Organ",'Region','Punch','Hemisphere')]
melt_mtcars$MAF <- sqrt(melt_mtcars$MAF)
casted_MAF_VarBySample <- dcast(melt_mtcars, CHROM.POS.REF.ALT~ID, value.var = "MAF" , na.rm=T)
rownames(casted_MAF_VarBySample) <- casted_MAF_VarBySample$CHROM.POS.REF.ALT
casted_MAF_VarBySample2 <- casted_MAF_VarBySample[ , !(names(casted_MAF_VarBySample) %in% c("CHROM.POS.REF.ALT"))]
```

```r
data[,'sample']<-rownames(data)
data2[,'sample']<-rownames(data2)
data3[,'sample']<-rownames(data3)
```

```r
data[,'sample']<-as.character(data[,'sample'])
data2[,'sample']<-as.character(data2[,'sample'])
data3[,'sample']<-as.character(data3[,'sample'])
```

```r
result <- apply(data, 2, sd)
result<-data.frame(na.omit(result))
result2 <- apply(data2, 2, sd)
result2<-data.frame(na.omit(result2))
result3 <- apply(data3, 2, sd)
result3 <-data.frame(na.omit(result3))
```

```r
results<-cbind(result,result2,result3)
colnames(results)<-c("DLX1","TBR1","COUPTF2")
```

```r
p <- ggplot(melt(results), aes(x=variable, y=value, color=variable)) + 
  geom_violin(binaxis='', stackdir='', dotsize=1) + scale_color_manual(values=c("#EF6666","#7CBD42","#61BDEA")) +
#geom_dotplot(binaxis='y', stackdir='center', dotsize=0.1)+
geom_jitter(shape=16, position=position_jitter(0.2)) + theme_classic()
```
