```r
library(pheatmap)
```

```r
data <- read.table("methylation_per_gene.tsv",header=TRUE, sep="\t")
```

```r
row_names = make.names(data[,1], unique = TRUE)
ready <- data[-1]
row.names(ready) <- row_names
```

```r
top_diff = scan("/projects/ps-gleesonlab7/User/arzoo/methylome_analysis/ex_in_diff_list_ascending_p_val.txt",what="",sep=",")
filtered <- ready[c("X8553_R_Br_sec3_Olig2.cutoff5",
                    "X8553_R_Br_sec3_TBR1.cutoff5",
                    "X8553_R_Br_sec3_DLX1.cutoff5",
                    "X8553_R_Br_sec3_SPI1.cutoff5",
                    "excitatory_200sc_control.cutoff5",
                    "inhibitory_200sc_control.cutoff5",
                    "heart_bulk_control.cutoff5"
                   )]
filtered1 <- filtered[top_diff[0:200],]
filtered1 <- filtered1[complete.cases(filtered1),]
filtered1_matrix <- as.matrix(filtered1)

heatmap(cosine(filtered1_matrix),scale="none",Rowv=FALSE, Colv=FALSE,margins=c(20,20))

pheatmap(filtered1_matrix,cluster_rows=FALSE)
```
